```python
import pandas as pd
import numpy as np

df = pd.read_excel(
    'bps.xls',
    header = 3,
    skipfooter = 2,
    na_values = ['-']
)
df
```

```python
df.rename(columns={'Unnamed: 0': 'Provinsi'}, inplace=True)
df = df.dropna()
df
```

```python
##mencari provinsi yg dimaksud
# min 1971
col=list(df.columns)
min_1971 = df[df[1971]== df[1971].min()]
prov_min_1971 = min_1971.values[0][0]
print('Provinsi dengan jumlah paling minimum pada tahun 1971 adalah', prov_min_1971)

# maks 2000
max_2000 = df[df[2000]==df[2000][:-1].max()]
prov_max_2000 = max_2000.values [0][0]
print('Provinsi dengan jumlah paling maksimum pada tahun 2000 adalah', prov_max_2000)


```

```python
#data indonesia


dataIndo=list(df.iloc[-1][1:])
dataIndo
```

```python
# data provinsi minimum 1971
minimum = df[df[1971] == df[1971].min()]
minimum

provinsi_min = list(minimum.values[0][1:])
provinsi_min
```

```python
# data provinsi maksimum 2000
maksimum = df[df[2000]== df[2000][:-1].max()]
maksimum

provinsi_maks = list(maksimum.values[0][1:])
provinsi_maks
```

```python
##plot grafik
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

bengkulu = df[df['Provinsi']=='Bengkulu']
jabar = df[df['Provinsi']=='Jawa Barat']
Indo = df[df['Provinsi']=='Indonesia']

th_minimum = list(bengkulu.columns)[1:]
th_maksimum = list(jabar.columns)[1:]
th_indo = list(Indo.columns)[1:]

plt.plot(
    th_minimum, provinsi_min,'r-',
    th_maksimum, provinsi_maks, 'b-',
    th_indo, dataIndo, 'y-'
    
)
plt.grid(True)
plt.legend(['Provinsi mimimum th 1970 (Bengkulu)','Provinsi maksimum th 2000 (Jawa Barat)','Populasi Indonesia'])
plt.show()
```
